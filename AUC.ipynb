```python
import pandas as pd
import numpy as np
from sklearn import metrics
# import seaborn as sns
import matplotlib.pylab as plt
```

```python
# 读取文件
# 读取数据文件
XGB1 = pd.read_csv('E:/cancerPrognosis/MyLUAD/code/forecast/XGBoost/resultGoup1.txt', sep="\t")
XGB2 = pd.read_csv('E:/cancerPrognosis/MyLUAD/code/forecast/XGBoost/resultGoup2.txt', sep="\t")
XGB3 = pd.read_csv('E:/cancerPrognosis/MyLUAD/code/forecast/XGBoost/resultGoup3.txt', sep="\t")

GBM1 = pd.read_csv('E:/cancerPrognosis/MyLUAD/code/forecast/lightGBM/resultGoup1.txt', sep="\t")
GBM2 = pd.read_csv('E:/cancerPrognosis/MyLUAD/code/forecast/lightGBM/resultGoup2.txt', sep="\t")
GBM3 = pd.read_csv('E:/cancerPrognosis/MyLUAD/code/forecast/lightGBM/resultGoup3.txt', sep="\t")

BPNN1 = pd.read_csv('E:/cancerPrognosis/MyLUAD/code/forecast/BPNN/resultGoup1.txt', sep="\t")
BPNN2 = pd.read_csv('E:/cancerPrognosis/MyLUAD/code/forecast/BPNN/resultGoup2.txt', sep="\t")
BPNN3 = pd.read_csv('E:/cancerPrognosis/MyLUAD/code/forecast/BPNN/resultGoup3.txt', sep="\t")
```

```python
# XGBoost 预测的三组结果的AUC值
plt.figure(num=None, figsize=(5,5), facecolor=None, edgecolor=None, clear=False, )
plt.rc('font', family='Times New Roman',size = 14)

XGB1_y = XGB1['y1'].tolist()
XGB1_pred = XGB1['pre1'].tolist()
fpr1, tpr1, thresholds = metrics.roc_curve(XGB1_y, XGB1_pred)
roc_auc1 = metrics.auc(fpr1, tpr1)
plt.plot(fpr1, tpr1, 'b', linewidth= 2, label='XGB1 = %0.2f' % roc_auc1)

XGB2_y = XGB2['y2'].tolist()
XGB2_pred = XGB2['pre2'].tolist()
fpr2, tpr2, thresholds = metrics.roc_curve(XGB2_y, XGB2_pred)
roc_auc2 = metrics.auc(fpr2, tpr2)
plt.plot(fpr2, tpr2, '#FF009D', linewidth= 2, label='XGB2 = %0.2f' % roc_auc2)

XGB3_y = XGB3['y3'].tolist()
XGB3_pred = XGB3['pre3'].tolist()
fpr3, tpr3, thresholds = metrics.roc_curve(XGB3_y, XGB3_pred)
roc_auc3 = metrics.auc(fpr3, tpr3)
plt.plot(fpr3, tpr3, 'g', linewidth= 2, label='XGB3 = %0.2f' % roc_auc3)

plt.legend(loc='lower right',fontsize = 12)
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([0, 1])  
plt.ylim([0, 1])  
plt.xlabel('1-Specificity',fontsize = 18)  
plt.ylabel('Sensitivity', fontsize = 18)  
# plt.title('Receiver operating characteristic example')
plt.show()
plt.savefig('E:/cancerPrognosis/MyLUAD/code/forecast/XGBoost/'+'XGB'+'.png')
```
